```python
def find_fear(text, word_list):
  words_set = set(word_list)

  text_list = text.split(" ")
  text_set = set(text_list)

  text_word_intersection = words_set.intersection(text_set)

  if len(text_word_intersection) > 0:
    return 1
  else:
    return 0
```
